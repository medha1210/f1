```python
%%sql 
SELECT r."driverId", d."forename", d."surname", COUNT(DISTINCT "raceId") AS race_count --Finding out the number of races Max and Lewis 
                                                                                       --have participated in so far
FROM "Race_Results" r
JOIN "Driver_Details" d ON r."driverId" = d."driverId"
WHERE r."driverId" IN (1, 830)
GROUP BY r."driverId", d."forename", d."surname";
```

```python
%%sql 
SELECT d."driverId", d."forename", d."surname", t."circuitId", t."circuitRef", MIN(l."time") AS best_lap_time --best lap time by tracks for lewis
FROM "Driver_Details" d
JOIN "Lap_Timings" l ON l."driverId" = d."driverId"
JOIN "Race_Schedule" rs ON rs."raceId" = l."raceId"
JOIN "Track_Information" t ON t."circuitId" = rs."circuitId"
WHERE d."driverId" = 1
GROUP BY d."driverId", d."forename", d."surname", t."circuitId", t."circuitRef"
ORDER BY t."circuitRef";
```

```python
%%sql 
SELECT d."driverId", d."forename", d."surname", t."circuitId", t."circuitRef", MIN(l."time") AS best_lap_time --best lap time by tracks for max
FROM "Driver_Details" d
JOIN "Lap_Timings" l ON l."driverId" = d."driverId"
JOIN "Race_Schedule" rs ON rs."raceId" = l."raceId"
JOIN "Track_Information" t ON t."circuitId" = rs."circuitId"
WHERE d."driverId" = 830
GROUP BY d."driverId", d."forename", d."surname", t."circuitId", t."circuitRef"
ORDER BY t."circuitRef";
```

```python
%%sql 
SELECT d."driverId", d."forename", d."surname", COUNT(*) AS total_wins, SUM (r."points") AS total_points -- total wins and points for hamilton
FROM "Driver_Details" d
JOIN "Race_Results" r ON d."driverId" = r."driverId"
WHERE r."positionOrder" = 1 AND d."driverId" = 1 
GROUP BY d."driverId", d."forename", d."surname";
```

```python
%%sql 
SELECT d."driverId", d."forename", d."surname", COUNT(*) AS total_wins, SUM (r."points") AS total_points --max's total wins and total points
FROM "Driver_Details" d
JOIN "Race_Results" r ON d."driverId" = r."driverId"
WHERE r."positionOrder" = 1 AND d."driverId" = 830
GROUP BY d."driverId", d."forename", d."surname";
```

```python
%%sql   
SELECT rs."year", d."driverId", d."forename", d."surname", AVG(SPLIT_PART(l."time", ':', 1)::NUMERIC * 60 + SPLIT_PART(l."time", ':', 2)::NUMERIC) AS avg_lap_time_seconds 
FROM "Driver_Details" d -- avg lap time for hamilton over the seasons
JOIN "Lap_Timings" l ON l."driverId" = d."driverId"
JOIN "Race_Schedule" rs ON rs."raceId" = l."raceId"
WHERE d."driverId" = 1
GROUP BY rs."year", d."driverId", d."forename", d."surname"  
ORDER BY rs."year";
```

```python
%%sql    
SELECT rs."year", d."driverId", d."forename", d."surname", AVG(SPLIT_PART(l."time", ':', 1)::NUMERIC * 60 + SPLIT_PART(l."time", ':', 2)::NUMERIC) AS avg_lap_time_seconds 
FROM "Driver_Details" d -- avg lap time for max over the seasons 
JOIN "Lap_Timings" l ON l."driverId" = d."driverId"
JOIN "Race_Schedule" rs ON rs."raceId" = l."raceId"
WHERE d."driverId" = 830
GROUP BY rs."year", d."driverId", d."forename", d."surname"  
ORDER BY rs."year";
```

```python
%%sql
SELECT 
    pole_driver."forename" || ' ' || pole_driver."surname" AS pole_position_driver, --who had pole position vs who won the race
    race_winner."forename" || ' ' || race_winner."surname" AS race_winner_driver,
    rs."year"
FROM "Race_Schedule" rs
JOIN "Qualifying_Results" qr 
    ON rs."raceId" = qr."raceId"
JOIN "Race_Results" rr 
    ON rs."raceId" = rr."raceId"
JOIN "Driver_Details" pole_driver 
    ON qr."driverId" = pole_driver."driverId"
JOIN "Driver_Details" race_winner 
    ON rr."driverId" = race_winner."driverId"
WHERE 
    qr."position" = 1  
    AND rr."positionOrder" = 1  
    AND rs."year" BETWEEN 2020 AND 2024  
ORDER BY rs."year";
```

```python
%%sql 
SELECT  td."constructorRef" ,SUM(cr."points") as total_points, rs."year" --total points for teams between 2000 and 2024
FROM "Team_Details" td 
JOIN "Constructor_Rankings" cr ON cr."constructorId" = td."constructorId"
JOIN "Race_Schedule" rs ON cr."raceId" = rs."raceId"
WHERE cr."constructorId" = td."constructorId" AND rs."year" BETWEEN 2000 AND 2024  
GROUP BY rs."year",td."constructorRef"
ORDER BY  rs."year" DESC, total_points DESC;
```

```python
%%sql 
SELECT 
    td."constructorRef" AS team_name, -- races participated in vs races qualitfied in team wise
    COUNT(DISTINCT qr."raceId") AS total_races_participated,
    COUNT(DISTINCT CASE WHEN qr."q3" IS NOT NULL THEN qr."raceId" END) AS times_qualified_for_q3
FROM "Team_Details" td
JOIN "Qualifying_Results" qr ON td."constructorId" = qr."constructorId"
GROUP BY td."constructorRef"
ORDER BY times_qualified_for_q3 DESC;
```

```python
%%sql
SELECT t."circuitRef", AVG(SPLIT_PART(l."time", ':', 1)::NUMERIC * 60 + SPLIT_PART(l."time", ':', 2)::NUMERIC) AS avg_lap_time_seconds, MIN(l."time") AS fastest_lap, MAX(l."time") AS slowest_lap
FROM "Lap_Timings" l --avg,slowest and fastest lap circuit wise
JOIN "Race_Schedule" rs ON rs."raceId" = l."raceId"
JOIN "Track_Information" t ON t."circuitId" = rs."circuitId"
GROUP BY t."circuitId", t."circuitRef"
ORDER BY avg_lap_time_seconds ASC;
```

```python
%%sql
SELECT 
    t."circuitRef", --avg lap time difference between p2 and p1 circuit wise 
    AVG(
        (SPLIT_PART(l2."time", ':', 1)::NUMERIC * 60 + SPLIT_PART(l2."time", ':', 2)::NUMERIC) 
        - (SPLIT_PART(l1."time", ':', 1)::NUMERIC * 60 + SPLIT_PART(l1."time", ':', 2)::NUMERIC)
    ) AS avg_P1_P2_diff
FROM "Lap_Timings" l1
JOIN "Race_Schedule" rs ON rs."raceId" = l1."raceId"
JOIN "Driver_Rankings" dr1 ON dr1."raceId" = rs."raceId" AND dr1."position" = 1
JOIN "Lap_Timings" l2 ON l2."raceId" = rs."raceId"
JOIN "Driver_Rankings" dr2 ON dr2."raceId" = rs."raceId" AND dr2."position" = 2
JOIN "Track_Information" t ON t."circuitId" = rs."circuitId"
WHERE l1."driverId" = dr1."driverId" 
  AND l2."driverId" = dr2."driverId"
GROUP BY t."circuitRef"
ORDER BY avg_P1_P2_diff DESC;

```

```python
%%sql
SELECT DISTINCT ON (t."circuitRef") t."circuitRef", l."driverId", d."forename", d."surname", rs."year", 
    SPLIT_PART(l."time", ':', 1)::NUMERIC * 60 + SPLIT_PART(l."time", ':', 2)::NUMERIC AS fastest_lap_time_seconds
FROM "Lap_Timings" l --fastest lap on each track
JOIN "Race_Schedule" rs ON rs."raceId" = l."raceId"
JOIN "Track_Information" t ON t."circuitId" = rs."circuitId"
JOIN "Driver_Details" d ON d."driverId" = l."driverId"
ORDER BY t."circuitRef", fastest_lap_time_seconds ASC;
```
